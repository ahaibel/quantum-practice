```python
# Create Array
import numpy as np
width=8
height=4
random_array = np.random.randint(0,7, size=(4,8))
print(random_array)
```

```python
# Create Normal Vector
flat_array = random_array.flatten()
magnitude = np.linalg.norm(flat_array)
normalized_array = flat_array / magnitude
normal_vector = normalized_array[:32]
print(normal_vector)
```

```python
# Import
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram
```

```python
# Encoding
num_qubits = len(normal_vector)
theta_values = 2 * np.arcsin(normal_vector)
qc = QuantumCircuit(num_qubits)
for i in range(num_qubits):
    qc.ry(theta_values[i], i)

# Draw
qc.draw()
```
